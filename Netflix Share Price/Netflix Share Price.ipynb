```python
# The 200-day moving average is considered especially significant in stock trading. As long as the
# 50-day moving average of a stock price remains above the 200-day moving average, the stock is generally 
# thought to be in a bullish trend. A crossover to the downside of the 200-day moving average is interpreted as bearish.
```

```python
netflix_df['Open'].plot(label = 'No Moving Average')
netflix_df['MV50'] = netflix_df['Open'].rolling(50).mean()
netflix_df['MV50'].plot(label = 'MV50')
netflix_df['MV200'] = netflix_df['Open'].rolling(200).mean()
netflix_df['MV200'].plot(label = 'MV200')
plt.legend()
plt.show()
```

###  Daily Per Change

```python
netflix_df['returns'] = (netflix_df['Close'] / netflix_df['Close'].shift(1)) - 1
```

```python
netflix_df.head()
```

```python
netflix_df['returns'].hist(bins=50)
plt.show()
```

```python
netflix_df['returns'].plot(kind = 'kde')
plt.show()
```

```python
netflix_df['returns'].plot(kind = 'box')
plt.show()
```

###  Comulative Return

```python
netflix_df['compulative return'] = (1 + netflix_df['returns']).cumprod()
```

```python
netflix_df.head()
```

```python
netflix_df['compulative return'].plot()
plt.title('Comutative Returns VS Date')
plt.show()
```

```python
import matplotlib.dates as mpdates
```

```python
dataset['Date'] = pd.to_datetime(dataset['Date'])
  
# apply map function
dataset['Date'] = dataset['Date'].map(mpdates.date2num)
  
# creating Subplots
fig, ax = plt.subplots()
  
# plotting the data
candlestick_ohlc(ax, dataset.values, width = 0.6,
                 colorup = 'green', colordown = 'red', 
                 alpha = 0.8)
  
# allow grid
ax.grid(True)
  
# Setting labels 
ax.set_xlabel('Date')
ax.set_ylabel('Price')
  
# setting title
plt.title('Prices For the Period 01-07-2020 to 15-07-2020')
  
# Formatting Date
date_format = mpdates.DateFormatter('%d-%m-%Y')
ax.xaxis.set_major_formatter(date_format)
fig.autofmt_xdate()
  
fig.tight_layout()
  
# show the plot
plt.show()
```
